### Creating a dataframe for Iris Virginica

```python
vi =ir[(ir.Classification=='Iris-virginica')]
vi.head()
```

```python
vi.describe()
```

### Ploting the outliers in parameters for Iris Virginica

```python
plt.figure(figsize=(10,8))
plt.title("Outliers in Iris Virginica",fontsize=15,color='g')
vi.boxplot();
```

### Scatter plot of parameters for Iris Virginica

```python
plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.scatter(vi['PL'],vi['PW'],s=10,marker='o',c='red')
plt.title('Scatter plot Petal Length Vs Petal width of Iris Virginica',fontsize=10)
plt.xlabel('Petal Length',fontsize=10)
plt.ylabel('Petal Width',fontsize=10)
plt.subplot(2,1,2)
plt.scatter(vi['SL'],vi['SW'],s=10,marker='o',c='blue')
plt.title('Scatter plot Sepal Length Vs Sepal Width of Iris Virginica',fontsize=10)
plt.xlabel('Sepal Length',fontsize=10)
plt.ylabel('Sepal Width',fontsize=10)

plt.tight_layout()
```
